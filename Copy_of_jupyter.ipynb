```python
a = 5
b = 9
c = 3

if a >= b and a >= c:
    print("Greatest number is:", a)
elif b >= a and b >= c:
    print("Greatest number is:", b)
else:
    print("Greatest number is:", c)
```

```python
num = 5

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
ch = 'A'

if ch.isupper():
    print("Uppercase letter")
elif ch.islower():
    print("Lowercase letter")
else:
    print("Not an alphabet")
```

```python
x = "7"

if x.isdigit():
    print("It is a number")
else:
    print("It is a character")
```
